#### `Understanding r"\s+"`
- r → Raw string (so backslashes are treated literally)
- \s → Matches any whitespace character, including:
- space ( )
- tab (\t)
- newline (\n)
- '+' → Means one or more

```python
df1["Description"] = df1["Description"].str.replace(r"\s+", " ", regex=True).str.strip()
```

##### `Adding Columns for further analysis`

```python
# convert to positive to negitive
df1['Quantity'] = df1['Quantity'].abs()
```

```python
# add total price
df1['TotalPrice'] = round(df1['Quantity'] * df1['UnitPrice'],2)
```

```python
df1['Year'] = df1['InvoiceDate'].dt.year
```

```python
df1['Quarter'] = df1['InvoiceDate'].dt.quarter
```

```python
df1['Month'] = df1['InvoiceDate'].dt.month_name()
```

```python
df1['Day'] = df1['InvoiceDate'].dt.day_name()
```

```python
df1['Weekend'] = df1['Day'].isin(['Saturday','Sunday'])
```

```python
df1['Weekend'] = df1['Weekend'].astype('int')
```

```python
def order_size(qty):
    if qty <= 10:
        return 'Small'
    elif qty <= 50:
        return 'Medium'
    else:
        return 'Large'

df1['OrderSize'] = df1['Quantity'].apply(order_size)
```
